```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import matplotlib as mpl
```

```python
ini = np.genfromtxt("Initial galaxy.txt")
a1 = np.genfromtxt("alpha=0.25.txt")
a2 = np.genfromtxt("alpha=0.5.txt")
a3 = np.genfromtxt("alpha=0.75.txt")
fi = np.genfromtxt("Final galaxy.txt")
```

```python
mpl.rcParams['font.family'] = 'Tahoma'
fig, ax = plt.subplots(1,5, figsize=(15,3), sharey=True)
norm = plt.cm.colors.LogNorm()
ax[0].imshow(ini, cmap='hot', vmin=0, vmax=500)
ax[0].set_title(r"$\alpha = 0$")
ax[0].set_ylabel(r"Y [$R_e$]")
ax[1].imshow(a3 , cmap='hot', vmin=0, vmax=500)
ax[1].set_title(r"$\alpha = 0.25$")

ax[2].imshow(a2 , cmap='hot', vmin=0, vmax=500)
ax[2].set_title(r"$\alpha = 0.5$")
ax[3].imshow(a1 , cmap='hot', vmin=0, vmax=500)
ax[3].set_title(r"$\alpha = 0.75$")
test = ax[4].imshow(fi, cmap='hot', vmin=0, vmax=500)
ax[4].set_title(r"$\alpha = 1$")
cbar_ax = fig.add_axes([0.92,0.15,0.02,0.7])
fig.colorbar(test, cax=cbar_ax, label="Relative surface brightness")
for a in ax:
    a.set_facecolor('black')
    a.invert_yaxis()
plt.setp(ax.flat, xlabel=r"X [$R_e$]")
plt.savefig("Galaxy evolution")
```
